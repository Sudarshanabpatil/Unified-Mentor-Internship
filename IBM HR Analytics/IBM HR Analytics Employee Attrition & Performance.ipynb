 ## Univariate Analysis


#### For Numerical Columns

```python
num = df.select_dtypes(exclude=object)
```

```python
num.head()
```

#### Histograms for Each Numerical Column
#####  histogram to understand the distribution of numerical columns

```python
for cols in num:
    plt.figure(figsize=(10, 5))
    # plt.subplot(1, 2, 1)
    sns.histplot(df[cols], kde=True)
    plt.title(f'Distribution of {cols}')
    plt.xlabel(cols)
    plt.ylabel('Frequency')
    plt.show()
```

#### Box Plots for Each Numerical Column
##### Box plots are useful for detecting outliers in each numerical column.

```python
for cols in num:
    # plt.figure(figsize=(10, 5))
    sns.boxplot(df[cols])
    plt.title(f'Distribution of {cols}')
    plt.xlabel(cols)
    plt.ylabel('Frequency')
    plt.show()
```

#### For Categorical Variables

```python
cat = df.select_dtypes(include=object)
cat.head()
```

##### Well visualize the counts of categorical variables such as Attrition, BusinessTravel

```python
count_of_attri = sns.countplot(x=df['Attrition'])
for cols in count_of_attri.containers:
    count_of_attri.bar_label(cols)
plt.show()
```

```python
count_business = sns.countplot(x=df['BusinessTravel'])
for cols in count_business.containers:
    count_business.bar_label(cols)
```

Pie chart For Department --->

```python
depart_count = df['Department'].value_counts()
plt.pie(depart_count,labels=depart_count.index,autopct='%1.1f%%',startangle=90)
plt.title('Department Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
count_educa = sns.countplot(x=df['EducationField'])
for i in count_educa.containers:
    count_educa.bar_label(i)
plt.show()    
```

```python
count_of_gender = sns.countplot(x=df['Gender'])
for cols in count_of_gender.containers:
    count_of_gender.bar_label(cols)
plt.show()
```

```python
plt.figure(figsize=(13, 4))
count_of_jobro = sns.countplot(x=df['JobRole'])
for cols in count_of_jobro.containers:
    count_of_jobro.bar_label(cols)
plt.xticks(rotation=45)
plt.show()
```
